```python
import pandas as pd
from matplotlib import pyplot as plt

CSV_FILE = '../data/asset.csv'
df = pd.read_csv(CSV_FILE,encoding="ISO-8859-1",parse_dates=['Date'])

plt.plot(df['Date'],df['Total'] - df['House'])

plt.gcf().autofmt_xdate() 

ax = plt.gca()
plt.minorticks_on()    
ax.yaxis.grid(which="minor", color='black', linestyle='dashed', linewidth=.3)
ax.yaxis.grid(which="major", color='black', linestyle='dashed', linewidth=.7)
ax.xaxis.grid(which="major", color='black', linestyle='dashed', linewidth=.3)
ax.set_yticklabels(['${:,.0f}K'.format(x/1000) for x in ax.get_yticks()])

plt.show()
```
